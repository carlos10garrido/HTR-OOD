```python
# Get for each dataset the n_grams frequencies until n = 5, so the final dictionary is freq_n_grams[dataset][n_gram] since later we will compare n_grams between datasets
def get_n_grams_frequencies(datasets):
    freq_n_grams = {}
    for dataset in datasets:
        freq_n_grams[dataset] = {}
        for n in range(1, 6):
            freq_n_grams[dataset][n] = {}
            n_grams = []
            for sentence in datasets[dataset]:
                # n_grams.extend(list(ngrams(sentence, n))) # Change to get the dict as continuous str instead of tuple
                # n_grams.extend([str(ngram) for ngram in list(ngrams(sentence, n))])
                n_grams.extend([''.join(ngram) for ngram in list(ngrams(sentence, n))])
            distr_n = nltk.FreqDist(n_grams)
            for key in distr_n:
                freq_n_grams[dataset][n][key] = (distr_n[key] / len(n_grams)) 

            if n <= 2:
                print(f'Finished dataset {dataset} for n = {n}. {freq_n_grams[dataset][n]}.')

    return freq_n_grams


freq_n_grams_train = get_n_grams_frequencies(train_datasets)
freq_n_grams_val = get_n_grams_frequencies(val_datasets)
freq_n_grams_test = get_n_grams_frequencies(test_datasets)
```

```python
# Compute KL-divergence between n-grams of different datasets. Always between source and target datasets and train and test splits

def compute_kl_divergence(freq_n_grams_train, freq_n_grams_test):
  kl_datasets = {}
  for dataset1 in freq_n_grams_train:
      kl_datasets[dataset1] = {}
      for dataset2 in freq_n_grams_test:
          kl_datasets[dataset1][dataset2] = {}
          for n in range(1, 6):
              kl_datasets[dataset1][dataset2][n] = {}
              for n_gram in freq_n_grams_train[dataset1][n]:
                  if n_gram in freq_n_grams_test[dataset2][n]:
                      kl_datasets[dataset1][dataset2][n][n_gram] = freq_n_grams_train[dataset1][n][n_gram] * np.log(freq_n_grams_train[dataset1][n][n_gram] / freq_n_grams_test[dataset2][n][n_gram])
                  else:
                      kl_datasets[dataset1][dataset2][n][n_gram] = freq_n_grams_train[dataset1][n][n_gram] * np.log(freq_n_grams_train[dataset1][n][n_gram] / 0.00000001)
              kl_datasets[dataset1][dataset2][n] = np.round(sum(kl_datasets[dataset1][dataset2][n].values()), 2)
  return kl_datasets
```

```python
kl_datasets = compute_kl_divergence(freq_n_grams_train, freq_n_grams_test)
kl_datasets
```

```python
# Get the average of the n-grams for each pair source-target

kl_datasets_avg = {}

for dataset1 in kl_datasets:
    kl_datasets_avg[dataset1] = {}
    for dataset2 in kl_datasets[dataset1]:
        kl_datasets_avg[dataset1][dataset2] = np.mean(list(kl_datasets[dataset1][dataset2].values()))

# kl_datasets_avg

# Put in pandas table

import pandas as pd

df = pd.DataFrame(kl_datasets_avg)
df

# Save as csv

df.to_csv('kl_divergence_datasets_FIXED.csv')
```
